```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("datafiles/datawhoo.xlsx")
df
```

```python
df['Date'] = pd.to_datetime(df['Date']) #convert to datetime format
#plotting the date
plt.plot(df['Date'], df['Sales'],color="Pink")
plt.title('Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.show()
         
```

```python

plt.bar(df['Product'], df['Sales'],color="gold")
plt.title('Sales Per Time')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.show()
```

```python
import numpy as np
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df['Max Price (Lakh)'],df['Insurance'])
m,b = np.polyfit(df['Max Price (Lakh)'],df['Insurance'],1)
plt.plot(df['Max Price (Lakh)'],m * df['Max Price (Lakh)']+b,color='blue',label='Regression line')
plt.title('Max Price (Lakh) Vs Insurance') 
plt.xlabel('Max Price (Lakh)')
plt.ylabel('Insurance')
plt.grid(True)
plt.show()
```

```python
df = pd.read_excel("datafiles/datawhoo.xlsx")
df.head()
```

```python
#Aggregate data by car model
sales_data = df.groupby('Category')['Sales'].sum()
sales_data
```

```python
#create the pie chart
plt.figure(figsize=(8,8))
plt.pie(sales_data, labels=sales_data.index, autopct='%1.1f%%')

plt.title('Proportion of Sales by Category')
plt.show()
```

```python
plt.hist(df['Sales'], bins=5,color='green',edgecolor='black')
#calculate center for each bin

plt.title('frequency per sale')
plt.xlabel('sales')
plt.ylabel('frequency')
plt.show()
```

```python
plt.fill_between(df['Date'],df['Sales'], color='yellow',alpha=0.7)
plt.plot(df['Date'], df['Sales'],color="Pink")
plt.title('Cumulative Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.show()

```
