```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("datafiles/Car data saved.xlsx")
df.head()
```

```python
df.columns
```

```python
cols=[ 'Min Price (Lakh)', 'Max Price (Lakh)', 'Range (kmpl)', 'CC',
       'Seats', 'Variants', 'Ex-Showroom Price', 'RTO', 'Insurance',
       'Other', 'Onroad Price']
```

```python
plt.figure(figsize=(15,12))
for i,var in enumerate(cols,start=1):
    plt.subplot(6,3,i)
    sns.scatterplot(x=df[var],y=df['Variants'],alpha=0.7)
    plt.title(f'Scatter plot of {var} vs onroad price')
    plt.xlabel(var)
    plt.ylabel('Target')
plt.tight_layout()
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(15,10))
for i,var in enumerate(cols,start=1):
    plt.subplot(3,4,i)
    sns.histplot(df[var],bins=20,color='orange', kde = True)
    plt.title(f'Histogram of {var}')
    plt.xlabel(var)
    plt.ylabel('Frequency')
plt.tight_layout()
plt.xticks(rotation=45)
plt.show()
```

```python
sns.boxplot(data=df['Other'])
plt.show()
```

```python
#create a boxplot of range by car
sns.boxplot(data=df, x='Type', y='Range (kmpl)',hue='Type',palette = ['yellow','purple'])
plt.title('Boxplot of Range by Car Type')
plt.show()
```

```python
selected_columns =['Min Price (Lakh)', 'Max Price (Lakh)', 'Range (kmpl)', 'CC',
       'Seats', 'Variants', 'Ex-Showroom Price', 'RTO', 'Insurance',
       'Other', ]
df_selected = df[selected_columns]
df_selected.head()

```

```python
corr = df_selected.corr()
corr
```

```python
sns.heatmap(corr, annot=True, cmap='Oranges')
plt.title('Heatmap of Correlation Matrix')
plt.show()
```

```python
#Violin plot of Max Price by Car Type
sns.violinplot(data=df,x='Type',y='Max Price (Lakh)',hue='Type',palette = ['yellow','purple'])
plt.title('Violin Plot of Maximum Price by Car Type')
plt.show()
```

```python
#Violin plot of Range by Car Type
sns.violinplot(data=df,x='Type', y='Range (kmpl)',hue='Type',palette = ['yellow','purple'])
plt.title('Violin Plot of Range by Car Type')
plt.show()
```
